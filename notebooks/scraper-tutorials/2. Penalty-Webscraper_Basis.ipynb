```python
!pip install scrapy
from scrapy import Selector
import requests
import pandas as pd
```

```python
url = "https://www.transfermarkt.de/bundesliga/elfmeterstatistik/wettbewerb/L1/saison_id/2017/plus/1"
html = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'}).content
page = Selector (text=html)
```

```python
### missed penalties
matchday = page.xpath("//div[@class = 'responsive-table'][1]//table[@class = 'items']//td[@class = 'zentriert'][1]/text()").extract()
penalty_taker = page.xpath("//div[@class = 'responsive-table'][1]//table[@class = 'items']//table[@class = 'inline-table']//td[@class = 'hauptlink']/a/@title").extract()
penalty_taker_position = page.xpath("//div[@class = 'responsive-table'][1]//table[@class = 'items']//table[@class = 'inline-table']/tr[2]/td/text()").extract()
goalkeeper = page.xpath("//div[@class = 'responsive-table'][1]//table[@class = 'items']//td[@class = 'links hauptlink no-border-links']/a/@title").extract()
score_before_penalty = page.xpath("//div[@class = 'responsive-table'][1]//table[@class = 'items']//td[@class = 'zentriert no-border-rechts no-border-links']/text()").extract()
minute = page.xpath("//div[@class = 'responsive-table'][1]//table[@class = 'items']//td[@class = 'zentriert'][2]/text()").extract()
home_team = page.xpath("//div[@class = 'responsive-table'][1]//td[@class = 'zentriert no-border-rechts'][3]/a/img/@alt").extract()
away_team = page.xpath("//div[@class = 'responsive-table'][1]//td[@class = 'zentriert no-border-links']/a/img/@alt").extract()
final_score = page.xpath("//div[@class = 'responsive-table'][1]//td[@class = 'zentriert no-border-rechts no-border-links']/a/span/text()").extract()
```

```python
general_url = "https://www.transfermarkt.de"
follow_links = [general_url + x for x in page.xpath("//div[@class = 'responsive-table'][1]//div[@class = 'pager']/ul//li[@class = 'page']/a/@href").extract()]
follow_links
```

```python
missed_statistics = list(zip(matchday, penalty_taker, penalty_taker_position, goalkeeper, score_before_penalty, minute, home_team, final_score, away_team)) 
df_missed = pd.DataFrame(missed_statistics, columns = ["matchday", "penalty_taker", "penalty_taker_position", "goalkeeper", "score_before_penalty", "minute", "home_team", "final_score", "away_team"])

df_missed.head(30)
```
