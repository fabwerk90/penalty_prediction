```python
!pip install scrapy
from scrapy import Selector
import requests
import pandas as pd
```

```python
url = "https://www.transfermarkt.de/spieler-statistik/wertvollstespieler/marktwertetop"
html = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'}).content
page = Selector (text=html)
```

```python
def printer(liste):
    print("All entries -", len(liste))
    print("------------------------")
    print(liste)
```

```python
### xpaths
player_table = page.xpath("//table[@class = 'items']")
```

```python
full_names = player_table.xpath("//table[@class = 'inline-table']/tr/td/a/img/@title").extract()

alt_names = player_table.xpath("//table[@class = 'inline-table']/tr/td/a/img/@alt").extract()

position = player_table.xpath("//table[@class = 'inline-table']/tr[2]/td/text()").extract()

age = player_table.xpath("//td[@class = 'zentriert'][2]/text()").extract()

nationality = player_table.xpath("//td[@class = 'zentriert'][3]/img[1]/@title").extract()

club = player_table.xpath("//td[@class = 'zentriert'][4]/a/img/@alt").extract()

market_value = player_table.xpath("//td[@class = 'rechts hauptlink']/b/text()").extract()
```

```python
data_lists = list(zip(full_names, position, age, nationality, club, market_value)) 
```

```python
df = pd.DataFrame(data_lists, columns = ['Name', 'Position', "Alter", "Nationalität", "Verein", "Marktwert"])

df.head(30)
```
